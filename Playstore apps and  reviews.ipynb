#Removing Duplicate values

```python
#checking duplicate values
df1_duplicate=df1.duplicated().sum()
df2_duplicate=df2.duplicated().sum()
print('The number of duplicates in Playstore_apps dataset and in Playstore_reviews dataset are :{} and  {} respectively' .format(df1_duplicate,df2_duplicate) )
```

```python
#drop duplicates
df1=df1.drop_duplicates()
df2=df2.drop_duplicates()
```

```python
#check for duplicates
df1_newduplicate=df1.duplicated().sum()
df2_newduplicate=df2.duplicated().sum()
print('Duplicates in playstore_apps file is:{}'.format(df1_newduplicate))
print('Duplicates in playstore_reviews file is :{}'.format(df2_newduplicate))
```
